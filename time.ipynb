```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aug = pd.read_csv('311_8-01_to_8-31function.csv')
```

```python
aug['created_date'] = pd.to_datetime(aug['created_date'])
aug["day_of_week"] = aug['created_date'].dt.day_name()
aug['closed_date'] = pd.to_datetime(aug['closed_date'])
aug['days_elapsed'] = (aug['closed_date'] - aug['created_date']).dt.days
aug['response_time'] = (aug['closed_date'] - aug['created_date'])
```

```python
aug['response_time_days'] = [i.total_seconds()/86400.0 for i in aug['response_time']]
aug['response_time_hours'] = [i.total_seconds()/3600.0 for i in aug['response_time']]
aug['response_time_minutes'] = [i.total_seconds()/60.0 for i in aug['response_time']]
aug['response_time_seconds'] = [i.total_seconds() for i in aug['response_time']] 
aug = aug[aug['days_elapsed'].notnull()]
```
